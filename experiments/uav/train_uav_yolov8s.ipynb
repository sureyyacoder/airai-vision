```python
from ultralytics import YOLO

#  hızlı eğitim için küçük model
model = YOLO("yolov8s.pt")


model.train(
    data="/content/drive/MyDrive/AirAi_Vision/processed/data_configs/uav_data.yaml",
    epochs=15,
    imgsz=640,
    batch=16,
    patience=5,
    name="yolov8s_uav_20ep_preview",
    project="/content/drive/MyDrive/AirAi_Vision/train_results",
    cache=True
)
```
