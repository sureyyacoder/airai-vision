```python
from ultralytics import YOLO

model_path = "/content/drive/MyDrive/AirAi_Vision/train_results/yolov8s_airplane_passenger_20ep/weights/best.pt"
model = YOLO(model_path)
```

```python
import os

# paths
val_images_dir = "/content/drive/MyDrive/AirAi_Vision/processed/images/valid/airplane/passenger"
val_labels_dir = "/content/drive/MyDrive/AirAi_Vision/processed/labels/valid/airplane/passenger"

# Boeing class index
BOEING_CLASS_ID = 1

#result list
missed_images = []

for filename in os.listdir(val_images_dir):
    if not filename.endswith(".jpg"):
        continue

    image_path = os.path.join(val_images_dir, filename)
    label_path = os.path.join(val_labels_dir, filename.replace(".jpg", ".txt"))

    # 1. Is there Boeing on the actual labels?
    has_true_boeing = False
    try:
        with open(label_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if int(parts[0]) == BOEING_CLASS_ID:
                    has_true_boeing = True
                    break
    except FileNotFoundError:
        continue

    # 2. If Boeing is on the labels â†’ Did the model guess it?
    if has_true_boeing:
        results = model.predict(image_path, conf=0.25, verbose=False)

        found_boeing_prediction = False
        for result in results:
            if result.boxes:
                for box in result.boxes:
                    if int(box.cls) == BOEING_CLASS_ID:
                        found_boeing_prediction = True
                        break
            if found_boeing_prediction:
                break

        if not found_boeing_prediction:
            missed_images.append(image_path)
            print(f" False Negative (Missed Boeing): {image_path}")
```
