```python
from ultralytics import YOLO

model = YOLO('yolov8m.pt')  # YOLOv8m = middle level model


model.train(
    data='/content/drive/MyDrive/AirAi_Vision/processed/data_configs/airplane_military_data.yaml',
    epochs=75,
    imgsz=640,
    batch=16,
    patience=20,
    name='yolov8m_airplane_military',
    project='/content/drive/MyDrive/AirAi_Vision/train_results',
    cache=True,
    device=0,
    verbose=True,
)
```
